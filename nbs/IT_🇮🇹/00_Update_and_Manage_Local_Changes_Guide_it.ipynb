# Aggiornare il Proprio Repository Locale Mantenendo le Modifiche Locali

Questa guida è destinata agli studenti che utilizzano il repository GitHub [Principi di Controlli Automatici](https://github.com/andreamunafo/principles-of-automatic-controls). È pensata per aiutarvi ad aggiornare il vostro repository locale con le ultime modifiche senza perdere le vostre personalizzazioni, in particolare all'interno della sottodirectory `tclab`, dove è incoraggiata la sperimentazione personale.

## Prerequisiti

Prima di procedere, assicuratevi di avere Git installato sul vostro computer e di avere una certa familiarità con l'uso della riga di comando o del terminale.

## Passo 1: Commit delle Modifiche Locali

Per evitare di perdere qualsiasi lavoro, iniziate facendo il commit di qualsiasi modifica abbiate fatto localmente. Questo è particolarmente rilevante per i quaderni nella directory `tclab` che avete modificato per le vostre sperimentazioni.

Aprite il vostro terminale o prompt dei comandi, navigate nella cartella del vostro repository locale ed eseguite i seguenti comandi:

```bash
git add .
git commit -m "Commit delle modifiche locali prima di recuperare gli aggiornamenti"
```

Questi comandi mettono in scena e fanno il commit delle vostre modifiche, preparando il vostro repository per ricevere gli aggiornamenti.

## Passo 2: Fetch e Merge degli Aggiornamenti

Invece di fare direttamente un pull (che potrebbe unire automaticamente le modifiche e portare a conflitti), farete un fetch degli aggiornamenti e poi li unirete manualmente. Questo approccio vi dà più controllo sull'integrazione dei nuovi aggiornamenti con le vostre modifiche.

Nel vostro terminale, all'interno della vostra cartella del repository locale, eseguite:

```bash
git fetch origin
git merge origin/master
```

**Nota:** Se il ramo principale del repository ha un nome diverso (ad esempio, `main` invece di `master`), regolate di conseguenza il comando.

## Passo 3: Risolvere Eventuali Conflitti di Merge

Se ci sono conflitti tra le vostre modifiche e gli aggiornamenti, Git interromperà il processo di merge e evidenzierà i file che necessitano di attenzione. Vedrete dei marcatori all'interno dei file in conflitto che indicano le sezioni che Git non è stato in grado di unire automaticamente.

Usate il vostro editor di testo o IDE preferito per aprire e risolvere questi conflitti. Dovrete modificare manualmente i file per integrare le modifiche o scegliere quale versione mantenere.

Dopo aver risolto i conflitti, contrassegnateli come risolti facendo lo staging e il commit:

```bash
git add .
git commit -m "Risolti i conflitti di merge"
```

## Passo 4: Push delle Modifiche (Non Consentito)

Conforme alle linee guida del repository, non vi è consentito fare il push di eventuali modifiche al repository principale. Pertanto, assicuratevi di mantenere le vostre modifiche locali, in particolare all'interno della sottodirectory `tclab`, dove state conducendo le vostre sperimentazioni.

## Linee Guida Aggiuntive

- **Non Modificare Notebooks Fuori da `tclab`:** Vi è richiesto di non modificare nessun notebook al di fuori della directory `nbs/tclab`. Questo garantisce che i materiali didattici di base rimangano consistenti per tutti gli studenti.
- **Aggiornamenti Frequenti:** Aggiornare regolarmente il vostro repository locale può aiutare a minimizzare potenziali conflitti e garantire di avere i materiali didattici più recenti.
- **Comunicare:** Se incontrate problemi significativi o avete bisogno di chiarimenti su come gestire specifici aggiornamenti, non esitate a chiedere orientamento.
